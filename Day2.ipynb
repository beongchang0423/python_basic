#튜플과 리스트의 소요시간 비교
튜플이 더 빠른 것을 볼 수 있다.  
그렇기에 변하지 않는 값들은 튜플을 사용하는 것이 좋다 

```python
import time
curr = time.time()
total = []
for i in range(1000):
    for j in range(100):
        total.append((i,j))
last = time.time()
print(last - curr)
```

```python
import time
curr= time.time()
total = []
for i in range(1000):
    for j in range(100):
        total.append([i,j])
last = time.time()
print(last - curr)
```

### 변수 여럿에 동시 할당하기 = unpacking
좌변과 우변의 개수가 같을때 한번에 할당하는 방식.
아래와 같은 입력은 튜플의 입력방식 중 하나로 튜플을 unpacking해서 값이 할당해주는 방식이다.

```python
s1,s2 = "text1","text2"
print(s1,s2)
```

```python
튜플 외에 다른 시퀀스 자료형도 unpacking이 가능하다
```

```python
s3,s4,s5 =["이렇게","리스트도","가능하다"]
print(s3,s4,s5)
```

### map() 내장 함수 이용하기  
iterable한 객체에서 요소 하나하나 꺼내어 특정 함수를 각각 적용하고 싶을 때 사용
반환은 map 클래스로 반환하는 iterbal한 객체이다.
이를 활용하기 위해 list()함수나 tuple()함수를 활용하여 map 클래스는 list나 tuple로 변경 가능하다.

```python
lst = [1.2,3.4,5.6,7.8,26.5,32]
lst = list(map(float,lst))
print(lst)
```

```python
print("value",1,2,3,"text")
print("value",1,2,3,"text",sep="\t")
```

```python
year = "2021"
month = "02"
day = "03"
hour = "18"
minute = "33"
second = "23"
print(year,month,day,sep="/",end=" ")
print(hour,minute,second,sep=":")
```
